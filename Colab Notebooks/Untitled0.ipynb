```
import numpy as np
import pandas as pd

# pd.set_option('display.max_rows', None)

players = pd.read_pickle('/content/drive/My Drive/Dashboard/Dashboard (Copy 1)/data/players_Clean.pkl')
market_value = pd.read_pickle('/content/drive/My Drive/Dashboard/Dashboard (Copy 1)/data/markval_Clean.pkl')
national_stats_full = pd.read_pickle('/content/drive/My Drive/Dashboard/Dashboard (Copy 1)/data/natstats_Clean.pkl')
national_stats = national_stats_full[0:100000]
# stats_full = pd.read_pickle('/content/stat_ds.pkl')
stats_full = pd.read_pickle('/content/drive/My Drive/Dashboard/Dashboard (Copy 1)/data/stats_Clean.pkl')
# stats = stats_full[0:100000]
transfers = pd.read_pickle('/content/drive/My Drive/Dashboard/Dashboard (Copy 1)/data/transfers_Clean.pkl')
trophies = pd.read_pickle('/content/drive/My Drive/Dashboard/Dashboard (Copy 1)/data/trophies_Clean.pkl')

def get_viz_df(stat_df, players_df, mv_df):

  players_df.reset_index(drop=True, inplace=True)
  terminate_players = players_df[players_df['age'].isna()].index
  players_df.drop(terminate_players, inplace=True)
  players_df.reset_index(drop=True, inplace=True)

  corr_df = pd.DataFrame(columns=None)
  
  stat_df.drop(columns=['club', 'competition'], inplace=True)
  stat_df.reset_index(drop=True, inplace=True)				
  stat_df['date'] = stat_df['date'].dt.strftime('%Y')
  term_index = stat_df[(stat_df['date'].isna())].index
  stat_df.drop(term_index, inplace=True)
  stat_df['date'] = stat_df['date'].astype(int)
  stat_df.reset_index(drop=True, inplace=True)
  term_index = stat_df[(stat_df['date']<2004) | (stat_df['date']>2019)].index
  stat_df.drop(term_index, inplace=True)
  stat_df.drop(columns=['date'], inplace=True)
  stat_df.groupby(['tm_id', 'season'])
  stat_sum_df = pd.pivot_table(stat_df, index=['tm_id', 'season'], columns='attribute', values='value', aggfunc=np.sum, fill_value=0)
  stat_mean_df = pd.pivot_table(stat_df, index=['tm_id', 'season'], columns='attribute', values='value', aggfunc=np.mean, fill_value=0)
  # stat_sum_df.columns = stat_df.columns.add_categories(['tm_id', 'season', 'id_season'])
  stat_sum_df['id_season'] = stat_df.index
  stat_sum_df.reset_index(drop=True, inplace=True)
  stat_sum_df[['tm_id', 'season']] = pd.DataFrame(stat_df['id_season'].tolist(), index=stat_df.index)
  stat_sum_df.drop(columns=['id_season'], inplace=True)
  # stat_mean_df.columns = stat_df.columns.add_categories(['tm_id', 'season', 'id_season'])
  stat_mean_df['id_season'] = stat_df.index
  stat_mean_df.reset_index(drop=True, inplace=True)
  stat_mean_df[['tm_id', 'season']] = pd.DataFrame(stat_df['id_season'].tolist(), index=stat_df.index)
  stat_mean_df.drop(columns=['id_season'], inplace=True)
  stat_sum_df['ppg'] = stat_mean_df['ppg'] / 100
  corr_df['tm_id'] = stat_sum_df['tm_id']
  corr_df[['app', 'a', 'g', 'mpg', 'cs', 'gc', 'mp', 'og', 'ppg', 'yc', 'syc', 'rc', 'son', 'soff', 's']]\
        = stat_sum_df[['app', 'a', 'g', 'mpg', 'cs', 'gc', 'mp', 'og', 'ppg', 'yc', 'syc', 'rc', 'son', 'soff', 's']]
  stat_df.set_index('tm_id',inplace=True)

  return corr_df

viz_df = get_viz_df(stats_full, players, market_value)

print(viz_df.head(100))
print(len(viz_df))
# print(viz_df.info())

```
