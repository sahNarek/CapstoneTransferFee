```
df = pd.read_pickle("/content/drive/My Drive/Capstone/Data/Clean/countries_leagues_Clean.pkl")
countries = pd.unique(df.country)
continets = []

for country in countries:
	if(country == 'England' or country == 'Bosnia-Herzegovina' or country == 'Serbia and Montenegro' or country == 'Scotland'\
		or country == 'Wales' or country == 'Northern Ireland'):
		continets.append('EU')
	elif(country == 'Botsuana' or country == 'Cote d\'Ivoire' or country == 'The Gambia' or country == 'DR Congo'):
		continets.append('AF')
	elif(country == 'Korea, North' or country == 'Korea, South' or country == 'Chinese Taipei (Taiwan)'):
		continets.append('AS')
	elif(country == 'Country'):
		continets.append('PG')
	else:
		country_code = pc.country_name_to_country_alpha2(country, cn_name_format="default")
		continent_name = pc.country_alpha2_to_continent_code(country_code)
		continets.append(continent_name)
```

```
df.set_index('country', inplace=True)
count_cont_df = pd.DataFrame(list(zip(countries, continets)), columns=['country', 'continent'])
count_cont_df.set_index('country', inplace=True)
df['continent'] = count_cont_df['continent']
df.reset_index(inplace=True)
# df.to_pickle('countries_leagues_continents_Clean.pkl')
```

```
df.continent.unique()
continents = (df >> 
 group_by(X.country) >> 
 summarise(country = X.country.unique()[0],
           continent = X.continent.unique()[0]))
continents.continent.unique()
```
