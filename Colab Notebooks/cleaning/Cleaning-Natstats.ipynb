```python
import pandas as pd
import numpy as np
import warnings
import re

warnings.filterwarnings('ignore')
```

```python
natstats = pd.read_csv("natstats_Full.csv", na_values = "-").fillna("0")
natstats
```

```python
def clean_stats_value(row):
    value = row.value
    if(row.Attribute == "Minutes played" or row.Attribute == "Minutes per goal"):
      value = int(str(value).replace('.', ""))
      print(value)
    elif(row.Attribute == "Appearances" or row.Attribute == "Squad"):
      value = int(str(value).replace(".0", ""))
      print(value)
    elif(row.Attribute == "PPG"):
      value = float(value)/100
      print(value)
    else:
      value = int(value)
    return value
```

```python
natstats.dtypes
```

```python
natstats["value"] = natstats['value'].str.replace('[^.\w\s]','')
natstats["value"] = natstats.apply(clean_stats_value, axis = 1)
natstats["value"] = pd.to_numeric(natstats["value"], errors = "coerce")
natstats.dtypes
```

```python
def check_string(column):
    col = column.str.contains('[@_!#$%^&*()<>?/\|}{~:]')
    idx = list(np.where(col == True)[0])
    rows = column.iloc[idx]
    
    if len(rows) == 0:
        return "Column does not contain any special characters"
    else:
        return rows
```

```python
check_string(natstats["Name"])
```

```python
natstats = natstats.drop(["Player_Id", "Name"], axis = 1)
natstats
```
