```
def season_to_date(season):
    date = season
    try:
      end = season.split("/")[1]
      date = "20" + str(end)
    except:
      pass
    return date

def add_months(date):
    result = date
    try:
       result = date + relativedelta(months=+5) + dt.timedelta(days=+29)
    except Exception as e:
        print(str(e))
        pass
    return result

def trophy_season(date):
  try:
    split_seasons = date.split("/")
    result = ""
    if(len(split_seasons) == 2):
      result = split_seasons[1] if split_seasons[1] != "" else split_seasons[0]
    else:
      result = split_seasons[0]
  except Exception as e:
    print(str(e))
  return result

trophy_season("2012/")

# "2012/".split("/")
```

```
def set_index_by_season(df_path):
  df = pd.read_pickle(df_path)
  df["Date"] = df.Season.apply(season_to_date)
  # df["Date"] = df.Date.astype('int')
  df["Date"] = pd.to_datetime(df.Date, format='%Y', errors = 'coerce')
  df["Date"] = df["Date"].apply(add_months)
  df["Date"] = pd.to_datetime(df.Date,format='%YYYY-%MM-%DD', errors = 'coerce')
  df.set_index("Date", drop=True, inplace=True)
  return df
```
