```
def short_position(position):
  result = ""
  if(position == "Goalkeeper"):
    result = "GK"
  elif(position == "Defender - Centre-Back"):
    result = "CB"
  elif(position == "Defender - Left-Back"):
    result = "LB"  
  elif(position == "Defender - Right-Back"):
    result = "RB"
  elif(position == "Midfielder - Defensive Midfield"):
    result = "CDM"
  elif(position == "Midfielder - Central Midfield"):
    result = "CM"
  elif(position == "Midfielder - Attacking Midfield"):
    result = "CAM"
  elif(position == "Forward - Left Winger"):
    result = "LW"
  elif(position == "Forward - Right Winger"):
    result = "RW"
  elif(position == "Forward - Second Striker"):
    result = "ST"
  elif(position == "Forward - Centre-Forward"):
    result = "CF"
  elif(position == "Midfielder - Left Midfield"):
    result = "LM"
  elif(position == "Midfielder - Right Midfield"):
    result = "RM"
  elif(position == "Forward"):
    result = "FW"
  elif(position == "Midfielder"):
    result = "MF"
  elif(position == "Defender"):
    result = "DF"
  return result
```

```
players["Position"] = players.Position.apply(short_position)
```
