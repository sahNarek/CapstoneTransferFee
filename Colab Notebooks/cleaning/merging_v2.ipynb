```
def tm_initial(name):
  result = ""
  split = name.split(" ")
  if(len(split) > 1):
    for part in split:
      result += part[0]
  else:
    result = split[0][0]
  return result.lower()

def sofifa_initial(name):
  result = ""
  if("." in name):
    split = name.split(".")
    for part in split:
      result += part[0]
  else:
    split = name.split(" ")
    for part in split:
      result += part[0]
  return result.lower()


def team_initial(name):
  result = ""
  name = name.lower()
  if("fc" in name):
    name = name.replace("fc", "")
  elif("cf" in name):
    name = name.replace("cf", "")
  elif("ac" in name):
    name = name.replace("ac", "")
  name = name.strip()
  split = name.split(" ")
  print(split)
  if(len(split) > 1):
    for part in split:
      result += part[0]
  else:
    result = split[0][0]
  return result.lower()


def find_sofifa_id(row):
  date_str = str(row.DOB)
  dob = date_str.replace("00:00:00","").strip()
  name = tm_initial(row.Name)
  try:
    sofifa_dob = sofifa[dob]
    name_cond = sofifa_dob.short_name.apply(lambda x: sofifa_initial((x)) == name)
    matches = sofifa_dob[name_cond]
    ids = matches.sofifa_id.unique()
  except Exception as e:
    ids = []
    pass
  ids = list(map(int, ids))
  return ids
```
