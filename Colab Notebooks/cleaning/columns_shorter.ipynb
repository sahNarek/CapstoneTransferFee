```
import pandas as pd
```

```
def shorter_columns(df):
  df.columns = [("_").join(col.lower().split(" ")) for col in df.columns]
  df.columns = [col.replace('team', 'club') for col in df.columns]
  return df
```

```
df = pd.read_pickle("/content/drive/My Drive/Capstone/Data/Clean/trophies_Clean.pkl")
df = shorter_columns(df)
df.columns
df.to_pickle("/content/drive/My Drive/Capstone/Data/Clean/trophies_Clean.pkl")
```
