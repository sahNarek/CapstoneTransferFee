* 2 = app / soff = substitutions off per game(qani xaxy meken zamen anum iran mtcnelov, erbvor inqy dashtum)
* 3 = goals / penalty goals = penalty goals per goal
* 4 = app / (mp / 90) = average games played on the field
* 5 = (`4` * ppg) / 3 = number of games won when the player played
* 6 = (`5`) / (s) = win percentage of the player
* 7 = squad / appearance = in the field per game(qani xaxy meka sastavum)
* 8 = mpg / 90 = goals per game
* 9 = mp / mpg = percentage of goal scoring time on the * pitch
* 10 = mp / yc = minutes per yellow card
* 11 = mp / rc = minutes per red card
* 12 = mp / a = minutes per assist
* 13 = mp / gc = minutes per conceded goal
* 14 = mp / cs = minutes per clean sheet
* 15 = cl / gc = goals conceded per one clean sheet
* 16 = `13` / 90 = games per conceded goal
* 17 = `14` / 90 = games per clean sheet
* 18 = yc / rc = yellow cards per red card
